```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_sales
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c'])

plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=0)

plt.show()
```

```python
plt.figure(figsize=(8,5))

ax = category_sales.plot(kind='bar')

for i, value in enumerate(category_sales):
    ax.text(i, value, f'{value:,.0f}', ha='center', va='bottom')

plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=0)

plt.show()
```

```python
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
category_profit
```

```python
plt.figure(figsize=(8,5))
category_profit.plot(kind='bar')

plt.title('Total Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.xticks(rotation=0)

plt.show()
```

```python
plt.figure(figsize=(8,5))

ax = category_profit.plot(
    kind='bar',
    color=['#4CAF50', '#2196F3', '#FF9800']
)

plt.title('Total Profit by Category', fontsize=14, fontweight='bold')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.xticks(rotation=0)

# Add value labels
for i, value in enumerate(category_profit):
    plt.text(i, value, f'{value:,.0f}', ha='center', va='bottom')

plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_sales
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

region_sales.plot(
    kind='bar',
    color=['#4CAF50', '#2196F3', '#FFC107', '#F44336']
)

plt.title('Total Sales by Region', fontsize=14)
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
ax = region_sales.plot(
    kind='bar',
    figsize=(8,5),
    color=['#4CAF50', '#2196F3', '#FFC107', '#F44336']
)

for i, value in enumerate(region_sales):
    ax.text(i, value, f'{value:,.0f}', ha='center', va='bottom')

plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)

plt.show()
```
